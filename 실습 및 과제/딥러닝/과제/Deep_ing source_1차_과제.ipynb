```python
import matplotlib.pyplot as plt
from pandas_datareader import data as pdr
import yfinance as yf
```

```python
# 데이터프레임

yf.pdr_override() # yfinance를 이용해서 pandas_datareader의 기능을 오버라이드(덮어쓰기)

samsung = pdr.get_data_yahoo('005930.KS', start='2022-02-08', end='2023-02-09', progress=False)
samsung
```

```python
# 데이터 시각화

plt.plot(samsung.index, samsung['Adj Close'])
```

```python
# 이동평균

# 30일 이동평균
ma30 = samsung['Adj Close'].rolling(window=30).mean()
samsung.insert(len(samsung.columns), "MA30", ma30)

# 90일 이동평균
ma90 = samsung['Adj Close'].rolling(window=90).mean()
samsung.insert(len(samsung.columns), "MA90", ma90)
```

```python
# 그래프 표시

plt.plot(samsung.index, samsung['Adj Close'], label="Adj Close")
plt.plot(samsung.index, samsung['MA30'], label="MA30")
plt.plot(samsung.index, samsung['MA90'], label="MA90")

plt.title(f"Stock Price Movement of SAMSUNG")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend(loc="best")
plt.grid(True)

plt.grid()
plt.show()
```

## 2. 사칙연산 함수

```python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    return a // b

def calculate(a, o, c):
    if o == "+":
        return add(a, c)
    elif o == "-":
        return subtract(a, c)
    elif o == "*":
        return multiply(a, c)
    elif o == "/":
        return divide(a, c)
    else:
        return False


expression = input().strip()
a, o, c = expression.split()
a, c = int(a), int(c)
if 1 <= a <= 100 and 1 <= c <= 100:
    result = calculate(a, o, c)
    print(f"{expression} = {result}")
else:
    print("Error: a and c must be between 1 and 100.")
```
