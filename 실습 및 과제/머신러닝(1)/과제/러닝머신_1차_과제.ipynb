# 2번 문제

```python
def plus(a, c):
    return a+c
def sub(a, c):
    return a-c
def divide(a, c):
    return a//c
def multi(a, c):
    return a*c
```

```python
input_data = input()
input_list = input_data.split()

a = int(input_list[0])
o = input_list[1]
c = int(input_list[2])
isFalse = False

if o == '+':
    result = plus(a, c)
elif o == '-':
    result = sub(a, c)
elif o == '/':
    result = divide(a, c)
elif o == '*':
    result = multi(a, c)
else:
    isFalse = True

if isFalse:
    print("False")
else:
    print(input_data + ' = ' + str(result))
```
