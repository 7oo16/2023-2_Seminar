```python
#20 이상이면 pass, 그렇지 않으면 fail 부여
mpg['test'] = np.where(mpg['total']>=20, 'pass', 'fail')
mpg.head()
```

```python
midwest = pd.read_csv('DAT/midwest.csv')
midwest.head()
```

```python
midwest.info()
```

```python
midwest_data = midwest.copy()
```

```python
midwest_data = midwest_data.rename(columns = {'poptotal':'total', 'popasian':'asian'})
midwest_data.head()
```

```python
midwest_data['아시아 인구 백분율'] = (midwest_data['asian']/midwest_data['total'])*100
midwest_data.head()
```

```python
asian_avg = midwest_data['아시아 인구 백분율'].mean()
asian_avg
```

```python
midwest_data['백분율 평균 비교'] = np.where(midwest_data['아시아 인구 백분율']>asian_avg, 'large', 'small')
midwest_data.head()
```

```python
count = midwest_data['백분율 평균 비교'].value_counts()
count
```

```python
count.plot.bar(rot=0)
```
